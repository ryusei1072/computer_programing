```python
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,#FeatureSet　の後を変えれば他の音響特徴量セットに変えることができる
    feature_level=opensmile.FeatureLevel.Functionals,
)
```

```python
#あるディレクトリにあるパス名を全て取得
pathlist = glob('./wav/para/*.wav')
pathlist = sorted(pathlist)
```

```python
#全ての音響特徴量のデータフレームを作成
df_para = DataFrame()
for i in range(len(pathlist)):
    feature = smile.process_file(pathlist[i])
    df_para = df_para.append(feature)
```
