```python
import tgt
```

```python
from pathlib import Path
```

```python
tgs = [tgt.read_textgrid(str(p)) for p in Path('./data').glob('*/*/*.TextGrid')]
```

```python
#　　　　　　x.text.split('@')[1]を削除したい
results = []
for tg in tgs:
    names = tg.get_tier_names()
    for name in names:
        try:
            results.extend([x.text.split('@')[1] for x in tgs[0].get_tier_by_name(name) if '@' in x.text])
        except Exception:
            pass
results
```
